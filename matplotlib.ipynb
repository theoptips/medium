```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.arange(0,10,0.1)
y_1 = np.array([1])*x_1
```

```python
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x_1,y_1, label="y=x") #can also specify color, marker, line style
#plt.plot(x_1, 6*x_1,label="y=3x")
#plt.plot(x_1, 1/3*x_1,label="y=1/3x")
plt.legend()
```

```python
# only show specific regions using plt.xlimit(0,1) plt.ylimit(0,1)
```
